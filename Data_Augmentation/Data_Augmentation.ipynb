Image Data Augmentation 방법
- rotation_range : 회전(0 ~ 360) -> **사용**

- width_shift_range = 좌, 우 이동(0 ~ 1)
- height_shift_range = 위, 아래 이동(0 ~ 1)
- shear_range = 찌그러뜨리기(0 ~ 1) -> **사용**
- zoom_range = 확대 -> **사용**
- horizontal_flip = True : 좌우 반전 -> **사용**
- vertical_flip = True : 상하 반전
- rescale : 조명 -> **사용**
- fill mode (원래 없었던 곳은 어떻게 채우는가?)
  - 'constant' : 검정색으로 채움 -> **사용**
  - 'nearest' : 가까운 것으로 채움 
  - 'reflect' : 거울로 채움
  - 'wrap' : 없어진 부분 -> 생겨난 부분


```
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img

datagen = ImageDataGenerator(
        rotation_range = 30,
        rescale = 1./255,
        shear_range = 0.2,
        zoom_range = 0.2,
        horizontal_flip = True,
        fill_mode='constant'
)
```

```
data_folder = "/content/gdrive/MyDrive/GraduationProject/data/cutAndMasking/maskedOutput/test/"
folder_list = os.listdir(data_folder)
for folder in folder_list:
  directory = data_folder + 'AUG/' + folder
  try:
    if not os.path.exists(directory):
      os.makedirs(directory)
  except OSError:
    print("Error: Creating Directory + directory")

  for filename in os.listdir(data_folder + folder):
    file = data_folder + '/' + folder + '/' + filename
    img = load_img(file, grayscale = True)
    x = img_to_array(img)
    x = x.reshape((1,) + x.shape)
    i = 0
    print(directory)
    for batch in datagen.flow(x, batch_size = 1, 
                            save_to_dir = directory,
                            save_prefix = folder, save_format = 'jpg'):
      i += 1
      if i > 10:
        break
```

완료된 후에 전체 데이터를 저장하는 방식인가?

- 학습 시에 이미지의 동일한 개수가 필요하다면 이후에 ex) 100개만 필요하다 -> 전체 데이터에서 shuffle 해서 100개만 가져오면 된다.
- 원본 데이터의 개수 -> 얼굴 인식 정확도를 연관지어 보는 것도 괜찮을 듯.
