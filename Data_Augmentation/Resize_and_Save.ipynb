```
data_folder = "/content/gdrive/MyDrive/GraduationProject/data/Train_Test/mask_test_AUG/"
folder_list = os.listdir(data_folder)
for folder in folder_list:
  directory = data_folder + 'Resized/' + folder
  try:
    if not os.path.exists(directory):
      os.makedirs(directory)
  except OSError:
    print("Error: Creating Directory + directory")
  
  cnt = 0
  for filename in os.listdir(data_folder + folder):
    file = data_folder + '/' + folder + '/' + filename
    image = cv2.imread(file)
    # try:
    #   imageResized = cv2.resize(image, (224, 224))
    # except:
    #   continue
    imageResized = cv2.resize(image, (224, 224))
    os.chdir(directory)
    filename = folder + str(cnt) + '.jpg'
    cv2.imwrite(filename, imageResized)
    print(filename)
    cnt += 1
```
